```python
import re

question = input("Ask Yourself!")
i = 0
X = []
print(question)

splited = re.split(r"\W+", question)
old_spl = splited.copy()

print(splited)

while i < len(splited):
    Y = []
    parent = splited[i]
    print(f"부모 = {parent}")
    splited.remove(parent)
    print(f"splited = {splited}")
    Y.append(parent)
    for j in splited:
        notes = input("take your mind!")
        print(f"notes = {notes}")
        Z = re.split(r"\W+", notes)
        Y.append([j, Z])
        print(f"기록 a = {Y}")
    X.append(Y)
    splited = old_spl.copy()
    i += 1

print(f"최종 : {X}")
```
