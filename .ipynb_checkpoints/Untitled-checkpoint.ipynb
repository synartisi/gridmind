```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xyz = [['a', ['b', ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ']], ['c', ['ㅅ', 'ㅇ', 'ㅈ']], ['d', ['ㅊ', 'ㅋ']], ['e', ['ㅌ', 'ㅍ', 'ㅎ']]], ['b', ['a', ['ㄱ', 'ㄴ', 'ㄷ']], ['c', ['ㄹ', 'ㅁ', 'ㅂ']], ['d', ['ㅅ', 'ㅇ', 'ㅈ', 'ㅊ']], ['e', ['ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']]], ['c', ['a', ['ㄱ']], ['b', ['ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅅ']], ['d', ['ㅇ', 'ㅈ', 'ㅊ', 'ㅋ']], ['e', ['ㅌ']]], ['d', ['a', ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ', 'ㅅ', 'ㅇ', 'ㅈ']], ['b', ['ㅊ']], ['c', ['ㅋ']], ['e', ['ㅌ']]], ['e', ['a', ['ㄱ', 'ㄴ', 'ㄷ', 'ㄹ', 'ㅁ', 'ㅂ']], ['b', ['ㅅ']], ['c', ['ㅇ', 'ㅈ']], ['d', ['ㅊ', 'ㅋ', 'ㅌ', 'ㅍ', 'ㅎ']]]]
print(xyz)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict

# 데이터
data = xyz

# 멀티 그래프 생성 (동일 노드 쌍에 여러 엣지)
G = nx.MultiDiGraph()

# 통계 저장
edge_stats = defaultdict(lambda: {'count': 0, 'comments': []})

for row in data:
    parent = row[0]
    
    for child_info in row[1:]:
        child = child_info[0]
        comments = child_info[1]
        
        # 각 코멘트를 개별 엣지로
        for comment in comments:
            G.add_edge(parent, child, comment=comment)
        
        # 통계
        edge_stats[(parent, child)]['count'] = len(comments)
        edge_stats[(parent, child)]['comments'] = comments

# 시각화 - 엣지 두께로 코멘트 개수 표현
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, k=3, iterations=50)

# 노드
nx.draw_networkx_nodes(G, pos, node_color='lightblue',
                       node_size=3000, edgecolors='black', linewidths=3)
nx.draw_networkx_labels(G, pos, font_size=20, font_weight='bold')

# 엣지 - 두께로 코멘트 개수 표현
for (parent, child), stats in edge_stats.items():
    width = stats['count'] * 2  # 코멘트 개수에 비례
    nx.draw_networkx_edges(G, pos, 
                          edgelist=[(parent, child)],
                          width=width,
                          edge_color='gray',
                          arrows=True,
                          arrowsize=20,
                          arrowstyle='->',
                          connectionstyle='arc3,rad=0.1',
                          alpha=0.7)
    
    # 라벨: 코멘트 개수
    mid_x = (pos[parent][0] + pos[child][0]) / 2
    mid_y = (pos[parent][1] + pos[child][1]) / 2
    plt.text(mid_x, mid_y, f"×{stats['count']}", 
            fontsize=14, fontweight='bold',
            bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8))

plt.title('멀티 그래프: 엣지 두께 = 코멘트 개수', fontsize=16, fontweight='bold')
plt.axis('off')
plt.tight_layout()
plt.show()

# 통계 출력
print("=== 엣지별 코멘트 통계 ===")
for (parent, child), stats in edge_stats.items():
    print(f"\n{parent} → {child}:")
    print(f"  코멘트 개수: {stats['count']}")
    print(f"  코멘트 목록: {stats['comments']}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 데이터를 플랫하게 펼치기
records = []

data = xyz

for row in data:
    parent = row[0]
    for child_info in row[1:]:
        child = child_info[0]
        comments = child_info[1]
        
        # 각 코멘트를 별도 레코드로
        for comment in comments:
            records.append({
                'parent': parent,
                'child': child,
                'comment': comment,
                'comment_length': len(comment)
            })

# DataFrame 생성
df = pd.DataFrame(records)
print("=== 플랫 데이터 ===")
print(df.head(10))

# 집계 분석
print("\n=== 부모-자식 쌍별 통계 ===")
stats = df.groupby(['parent', 'child']).agg({
    'comment': ['count', 'first', 'last'],  # 개수, 첫번째, 마지막
    'comment_length': ['mean', 'sum', 'min', 'max']
}).round(2)
print(stats)

# 코멘트 개수 행렬
print("\n=== 코멘트 개수 행렬 ===")
count_matrix = df.groupby(['parent', 'child']).size().unstack(fill_value=0)
print(count_matrix)

# 히트맵 시각화
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 1. 코멘트 개수
sns.heatmap(count_matrix, annot=True, fmt='d', cmap='YlOrRd',
           square=True, linewidths=2, ax=axes[0],
           cbar_kws={'label': '코멘트 개수'})
axes[0].set_title('코멘트 개수', fontsize=14, fontweight='bold')
axes[0].set_xlabel('자식')
axes[0].set_ylabel('부모')

# 2. 평균 코멘트 길이
avg_length_matrix = df.groupby(['parent', 'child'])['comment_length'].mean().unstack(fill_value=0)
sns.heatmap(avg_length_matrix, annot=True, fmt='.1f', cmap='Blues',
           square=True, linewidths=2, ax=axes[1],
           cbar_kws={'label': '평균 길이'})
axes[1].set_title('평균 코멘트 길이', fontsize=14, fontweight='bold')
axes[1].set_xlabel('자식')
axes[1].set_ylabel('부모')

# 3. 총 코멘트 길이
total_length_matrix = df.groupby(['parent', 'child'])['comment_length'].sum().unstack(fill_value=0)
sns.heatmap(total_length_matrix, annot=True, fmt='d', cmap='Greens',
           square=True, linewidths=2, ax=axes[2],
           cbar_kws={'label': '총 길이'})
axes[2].set_title('총 코멘트 길이', fontsize=14, fontweight='bold')
axes[2].set_xlabel('자식')
axes[2].set_ylabel('부모')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 최대 코멘트 개수 찾기
max_comments = 0
for row in data:
    for child_info in row[1:]:
        max_comments = max(max_comments, len(child_info[1]))

print(f"최대 코멘트 개수: {max_comments}")

# 3D 텐서 생성 (부모 × 자식 × 코멘트_인덱스)
nodes = ['1', '2', '3', '4']
n_nodes = len(nodes)

# 코멘트를 숫자로 인코딩 (예: 알파벳 순서)
def encode_comment(comment):
    """코멘트를 숫자로 변환 (예시)"""
    return sum(ord(c) for c in comment)

# 텐서 생성
comment_tensor = np.zeros((n_nodes, n_nodes, max_comments))

for row in data:
    parent = row[0]
    parent_idx = nodes.index(parent)
    
    for child_info in row[1:]:
        child = child_info[0]
        child_idx = nodes.index(child)
        comments = child_info[1]
        
        for i, comment in enumerate(comments):
            comment_tensor[parent_idx, child_idx, i] = encode_comment(comment)

print("\n=== 텐서 형태 ===")
print(f"Shape: {comment_tensor.shape}")
print(f"(부모 {n_nodes} × 자식 {n_nodes} × 최대코멘트 {max_comments})")

# 슬라이스 시각화
fig, axes = plt.subplots(1, max_comments, figsize=(5*max_comments, 5))

for i in range(max_comments):
    ax = axes[i] if max_comments > 1 else axes
    slice_data = comment_tensor[:, :, i]
    
    im = ax.imshow(slice_data, cmap='viridis', aspect='auto')
    ax.set_title(f'코멘트 인덱스 {i}', fontsize=12, fontweight='bold')
    ax.set_xlabel('자식')
    ax.set_ylabel('부모')
    ax.set_xticks(range(n_nodes))
    ax.set_yticks(range(n_nodes))
    ax.set_xticklabels(nodes)
    ax.set_yticklabels(nodes)
    
    # 값 표시
    for pi in range(n_nodes):
        for ci in range(n_nodes):
            value = slice_data[pi, ci]
            if value > 0:
                ax.text(ci, pi, f'{int(value)}', 
                       ha='center', va='center', 
                       color='white', fontweight='bold')
    
    plt.colorbar(im, ax=ax)

plt.suptitle('3D 텐서: 각 슬라이스 = 코멘트 인덱스', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()

# 텐서 집계
print("\n=== 텐서 분석 ===")
print(f"각 관계의 코멘트 개수 (0이 아닌 값):")
comment_counts = np.sum(comment_tensor != 0, axis=2)
print(comment_counts)
```

```python
import sqlite3
import pandas as pd

# 인메모리 DB 생성
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# 테이블 생성
cursor.execute('''
CREATE TABLE relations (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    parent TEXT,
    child TEXT,
    comment TEXT,
    comment_index INTEGER
)
''')

# 데이터 삽입
for row in data:
    parent = row[0]
    for child_info in row[1:]:
        child = child_info[0]
        comments = child_info[1]
        
        for idx, comment in enumerate(comments):
            cursor.execute('''
                INSERT INTO relations (parent, child, comment, comment_index)
                VALUES (?, ?, ?, ?)
            ''', (parent, child, comment, idx))

conn.commit()

# 쿼리 예제들
print("=== SQL 쿼리 분석 ===\n")

# 1. 부모별 총 코멘트 개수
print("1. 부모별 총 코멘트 개수:")
df1 = pd.read_sql_query('''
    SELECT parent, COUNT(*) as total_comments
    FROM relations
    GROUP BY parent
    ORDER BY total_comments DESC
''', conn)
print(df1)

# 2. 부모-자식 쌍별 코멘트 개수
print("\n2. 부모-자식 쌍별 코멘트 개수:")
df2 = pd.read_sql_query('''
    SELECT parent, child, COUNT(*) as comment_count
    FROM relations
    GROUP BY parent, child
    ORDER BY comment_count DESC
''', conn)
print(df2)

# 3. 특정 부모의 모든 관계
print("\n3. 부모 '1'의 모든 관계:")
df3 = pd.read_sql_query('''
    SELECT child, comment, comment_index
    FROM relations
    WHERE parent = '1'
    ORDER BY child, comment_index
''', conn)
print(df3)

# 4. 가장 많은 코멘트를 받는 자식
print("\n4. 가장 많은 코멘트를 받는 자식:")
df4 = pd.read_sql_query('''
    SELECT child, COUNT(*) as received_comments
    FROM relations
    GROUP BY child
    ORDER BY received_comments DESC
''', conn)
print(df4)

# 시각화
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 부모별
df1.plot(kind='bar', x='parent', y='total_comments', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('부모별 총 코멘트 개수', fontweight='bold')
axes[0,0].set_xlabel('부모')
axes[0,0].set_ylabel('코멘트 개수')

# 자식별
df4.plot(kind='bar', x='child', y='received_comments', ax=axes[0,1], color='lightcoral')
axes[0,1].set_title('자식별 받은 코멘트 개수', fontweight='bold')
axes[0,1].set_xlabel('자식')
axes[0,1].set_ylabel('코멘트 개수')

# 관계별
df2_pivot = df2.pivot(index='parent', columns='child', values='comment_count').fillna(0)
import seaborn as sns
sns.heatmap(df2_pivot, annot=True, fmt='.0f', cmap='YlGnBu', ax=axes[1,0])
axes[1,0].set_title('부모-자식 코멘트 개수 히트맵', fontweight='bold')

# 분포
all_counts = df2['comment_count'].values
axes[1,1].hist(all_counts, bins=10, color='mediumpurple', edgecolor='black')
axes[1,1].set_title('코멘트 개수 분포', fontweight='bold')
axes[1,1].set_xlabel('코멘트 개수')
axes[1,1].set_ylabel('빈도')

plt.tight_layout()
plt.show()

conn.close()
```

```python
from collections import Counter
import matplotlib.pyplot as plt

# 모든 코멘트 수집
all_comments = []
comment_by_relation = {}

for row in data:
    parent = row[0]
    for child_info in row[1:]:
        child = child_info[0]
        comments = child_info[1]
        
        all_comments.extend(comments)
        comment_by_relation[(parent, child)] = comments

# 1. 코멘트 빈도 분석
comment_freq = Counter(all_comments)
print("=== 코멘트 빈도 ===")
for comment, freq in comment_freq.most_common(10):
    print(f"{comment}: {freq}회")

# 2. 코멘트 길이 분석
comment_lengths = [len(c) for c in all_comments]
print(f"\n=== 코멘트 길이 통계 ===")
print(f"평균: {np.mean(comment_lengths):.2f}")
print(f"중앙값: {np.median(comment_lengths):.2f}")
print(f"최소: {min(comment_lengths)}, 최대: {max(comment_lengths)}")

# 3. 관계별 코멘트 다양성
print(f"\n=== 관계별 코멘트 다양성 ===")
for (parent, child), comments in comment_by_relation.items():
    unique_ratio = len(set(comments)) / len(comments) if comments else 0
    print(f"{parent}→{child}: {len(comments)}개 (유일성: {unique_ratio:.2%})")

# 시각화
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 빈도
top_comments = comment_freq.most_common(15)
axes[0,0].barh([c[0] for c in top_comments], [c[1] for c in top_comments], color='teal')
axes[0,0].set_xlabel('빈도')
axes[0,0].set_title('상위 15개 코멘트 빈도', fontweight='bold')
axes[0,0].invert_yaxis()

# 길이 분포
axes[0,1].hist(comment_lengths, bins=20, color='orange', edgecolor='black')
axes[0,1].set_xlabel('코멘트 길이')
axes[0,1].set_ylabel('빈도')
axes[0,1].set_title('코멘트 길이 분포', fontweight='bold')

# 관계별 개수
relations = list(comment_by_relation.keys())
counts = [len(comments) for comments in comment_by_relation.values()]
axes[1,0].bar(range(len(relations)), counts, color='purple', alpha=0.7)
axes[1,0].set_xticks(range(len(relations)))
axes[1,0].set_xticklabels([f"{p}→{c}" for p, c in relations], rotation=45)
axes[1,0].set_ylabel('코멘트 개수')
axes[1,0].set_title('관계별 코멘트 개수', fontweight='bold')

# 유일성
uniqueness = [len(set(comments))/len(comments) if comments else 0 
              for comments in comment_by_relation.values()]
axes[1,1].bar(range(len(relations)), uniqueness, color='green', alpha=0.7)
axes[1,1].set_xticks(range(len(relations)))
axes[1,1].set_xticklabels([f"{p}→{c}" for p, c in relations], rotation=45)
axes[1,1].set_ylabel('유일성 비율')
axes[1,1].set_title('관계별 코멘트 유일성', fontweight='bold')
axes[1,1].axhline(y=1.0, color='red', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
